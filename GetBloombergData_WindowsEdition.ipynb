```python
import pdblp
import pandas as pd
import numpy as np
import subprocess
```

```python
con = pdblp.BCon(debug=False, port=8194)
con = pdblp.BCon(timeout=5000)
con.start()
```

```python
banks_df = pd.read_csv('US_Active_Banks.csv')
bloomberg_ticker = banks_df['BloombergTicker']
news_and_sentiment_flds = pd.read_excel('bloomberg_data_flds.xlsx',sheet_name=0)
rating = pd.read_excel('bloomberg_data_flds.xlsx',sheet_name=1)
start_date = '20000101'
end_date = '20200211'
parent_path = 'banks'
```

```python
for ticker in bloomberg_ticker:
    ticker_safe_name = ticker.replace(' ','_')
    folder = '{parent}\{ticker}'.format(parent=parent_path,ticker=ticker_safe_name)
    !if not exist $folder mkdir $folder
    for fld in news_and_sentiment_flds['Fields']:
        file_name = ticker_safe_name+'_'+fld+'.csv'
        df = con.bdh(tickers=ticker,flds=fld,start_date=start_date,end_date=end_date)
        df.to_csv('{folder}\{file_name}'.format(folder=folder,file_name=file_name))
    print('Done for {ticker}'.format(ticker=ticker))
```
